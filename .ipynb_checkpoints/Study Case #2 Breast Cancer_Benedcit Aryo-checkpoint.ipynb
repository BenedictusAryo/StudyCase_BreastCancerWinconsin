### Training Dataset Preparation

```python
X=np.array(df.drop(columns=['diagnosis']))
y = np.array(df['diagnosis']=="M").astype(int)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
```

## Machine Learning Model

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
```

```python
model_1=LogisticRegression()
model_2=SVC()
model_3=DecisionTreeClassifier()
```

```python
print("X_train size : ",X_train.shape)
print("y_train size : ",y_train.shape)
```

```python
model_1.fit(X_train,y_train)
model_2.fit(X_train,y_train)
model_3.fit(X_train,y_train)
```

```python
print("Score Model 1 : ",model_1.score(X_test,y_test))
print("Score Model 2 : ",model_2.score(X_test,y_test))
print("Score Model 3 : ",model_3.score(X_test,y_test))
```
